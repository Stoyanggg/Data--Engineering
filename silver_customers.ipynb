```python
from pyspark.sql.functions import *
```

```python
# Load the "customers" Delta table into a Spark DataFrame

customers = spark.read.format("delta").table("stoyan.bronze_customers")
customers.display()
```

```python
# Convert column names to lowercase and replace '.' with '_' for standardization
new_col_cust = []                                                       # Initialize an empty list for new column names
for column in customers.columns:
    new_col_cust.append(column.lower().replace('.','_'))                # Convert to lowercase and replace '.' with '_'
customers = customers.toDF(*new_col_cust)                               # Rename columns in the DataFrame
customers.display()
```

```python
# Define a transformation function to process the "customers" DataFrame
def transformation (customers):
    # Standardize the format of phone numbers by replacing spaces with dashes
    customers= customers.withColumn('phone_number', regexp_replace('phone_number',' ', '-'))
    # Convert "date_of_birth" column from string to date format
    customers= customers.withColumn('date_of_birth', to_date(col('date_of_birth'),'dd-MMM-yy'))
    # Rename "ingest_datetime" column to "consume_timestamp" for clarity
    customers = customers.withColumnRenamed('ingest_datetime', 'consume_timestamp')
    # Drop the "batch_id" column as it is no longer required
    customers = customers.drop(col('batch_id'))
    # Add a new column "load_tsp" with the current timestamp to track loading time
    customers = customers.withColumn('load_tsp', current_timestamp())
    # Adjust "date_of_birth" for dates with years beyond 2000 (e.g., incorrect parsing of birth year)
    customers=customers.withColumn('date_of_birth',
                                 expr("IF(year(date_of_birth) > 2000, date_add(date_of_birth, -36525), date_of_birth)"))
    return customers                                # Return the transformed DataFrame
transformed_customers = transformation(customers)
transformed_customers.display()
```
